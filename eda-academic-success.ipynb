```python
# Define the columns and titles to be plotted
columns = ['marital_status', 'application_order','course',
       'attendance_type', 'displaced', 'special_needs', 'debtor',
       'fees_up_to_date', 'gender', 'scholarship_holder',
       'international_student', 'unemployment_rate',
       'inflation_rate', 'gdp']
titles = ['marital_status', 'application_order', 'course',
       'attendance_type', 'displaced', 'special_needs', 'debtor',
       'fees_up_to_date', 'gender', 'scholarship_holder',
       'international_student', 'unemployment_rate',
       'inflation_rate', 'gdp']

plt.figure(figsize=(25, 80))

for i, column in enumerate(columns):
    # Pie chart
    plt.subplot(15, 2, 2*i + 1)
    value_counts = df_train[column].value_counts()
    color_map = ['#ff3399','#003366','#ff9933']
    myexplode = [0.1] + [0.1] * (len(value_counts) - 1)
    plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', startangle=90, explode = myexplode)
    plt.title(f'{titles[i]} Distribution')
    
    # Count plot
    plt.subplot(15, 2, 2*i + 2)
    ax = sns.countplot(data=df_train, x=column, hue='target')
    for count in ax.containers:
        ax.bar_label(count)
    plt.title(f'{titles[i]} vs Target')

plt.tight_layout()
plt.show()
```

```python
# Separate features and target
X = df_train.drop('target', axis=1)
y = df_train['target']

# Initialize SMOTE
smote = SMOTE(random_state=42)

# Initialize the Random Forest Classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)

# Create a pipeline with SMOTE and the classifier
pipeline = Pipeline([('smote', smote), ('scaler', StandardScaler()), ('classifier', clf)])
```

```python
skf = StratifiedKFold(n_splits=5)
scorer = make_scorer(accuracy_score)
scores = cross_val_score(pipeline, X, y, cv=skf, scoring=scorer)
```

```python
print(f'Cross-Validation Accuracy Scores: {scores}')
print(f'Mean Cross-Validation Accuracy: {scores.mean()}')

pipeline.fit(X, y)
y_pred = pipeline.predict(X)
report = classification_report(y, y_pred, target_names=['Graduate', 'Dropout', 'Enrolled'])

print('Classification Report:')
print(report)
```
